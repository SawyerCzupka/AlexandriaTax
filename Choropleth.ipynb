```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("data/map_no_alex.geojson")
gdf
```

```python
import json

# Specify the file path
file_path = 'data/styledict.json'

# Load the styledict object from the JSON file
with open(file_path, 'r') as file:
    styledict = json.load(file)
```

# Folium Plot

```python
import logging
logging.basicConfig(level=logging.INFO)

import folium
from folium.plugins import TimeSliderChoropleth

m = folium.Map([38.8,-77.1], zoom_start=12)

TimeSliderChoropleth(
    gdf.to_json(),
    styledict = styledict,
).add_to(m)

m
```
